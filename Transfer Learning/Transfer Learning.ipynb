```python
img_generator = ImageDataGenerator(rescale=1.0/255)
```

```python
train = img_generator.flow_from_directory('/content/training_set/training_set',
                                          target_size=(224, 224),
                                          classes=['cats', 'dogs'],
                                          class_mode='binary'
                                          )

test = img_generator.flow_from_directory('/content/test_set/test_set',
                                          target_size=(224, 224),
                                          classes=['cats', 'dogs'],
                                         class_mode='binary'
                                          )
```

# Making model

```python
model = keras.applications.mobilenet_v2.MobileNetV2(
    input_shape=(224, 224, 3),
    weights='imagenet',
    include_top=False
)
```

```python
model.summary()
```

```python
# freeing the model
model.trainable = False
```

```python
# defining the custome head
model.output
```

```python
global_avarage_layer = keras.layers.GlobalAveragePooling2D()(model.output)
```

```python
global_avarage_layer
```

```python
# output layer and prediction layers
prediction_layers = keras.layers.Dense(units=1, activation='sigmoid')(global_avarage_layer)
```

```python
model_ = keras.models.Model(inputs=model.input, outputs=prediction_layers)
```

```python
model_.summary()
```

```python
# compile the model
opt = tf.keras.optimizers.RMSprop(
    learning_rate=0.001)
```

```python
model_.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# training the model

model_.fit(train, epochs=5, validation_data=test)
```

```python
# fine tuning

len(model.layers)
```
