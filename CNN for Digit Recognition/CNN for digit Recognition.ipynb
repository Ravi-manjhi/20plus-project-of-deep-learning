# Step 01 - Import statements

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras

import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score, confusion_matrix
```

# Data Preprocessing

## gether data

```python
from keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape, x_test.shape
```

```python
plt.figure(figsize=(2, 2))
plt.imshow(x_train[0])
plt.show()
```

```python
# smplifing data
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
plt.figure(figsize=(2, 2))
plt.imshow(x_train[1])
plt.show()
```

```python
x_train.shape, x_test.shape
```

```python
# reshape datasets
x_train = x_train.reshape(60000, 28, 28, 1)
x_test = x_test.reshape(10000, 28, 28, 1)
```

# Step 3 - Building the Model and Evaluation

```python
model = keras.models.Sequential()
```

```python
# addiing the layers
model.add(keras.layers.Conv2D(filters=32, kernel_size=3, padding='same',activation='relu', input_shape=(28, 28, 1)))

# maxpool layer
model.add(keras.layers.MaxPool2D(pool_size=(2, 2), strides=2, padding='valid'))

# adding second layers
model.add(keras.layers.Conv2D(filters=32, kernel_size=3, padding='same',activation='relu'))

# maxpool layer
model.add(keras.layers.MaxPool2D(pool_size=(2, 2), strides=2, padding='valid'))

# dropout
model.add(keras.layers.Dropout(0.4))

# flatting layers
model.add(keras.layers.Flatten())
```

```python
# fully connection layers
model.add(keras.layers.Dense(units=64, activation="relu"))

# output layers
model.add(keras.layers.Dense(units=10, activation='softmax'))
```

```python
model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
```

```python
# training the model
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

### Evaluation

```python
y_pred = model.predict(x_test)
pred = np.argmax(y_pred, axis=1)
```

```python
print("the accuracy of model on test dataset is:", accuracy_score(y_test, pred))
```

```python
confusion_matrix(y_test, pred)
```

# step 05 - Learinig curve

```python
def learning_curve(history):
  plt.plot(history.epoch, history.history['sparse_categorical_accuracy'])
  plt.plot(history.epoch, history.history['val_sparse_categorical_accuracy'])
  plt.title('Model Accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()

  plt.plot(history.epoch, history.history['loss'])
  plt.plot(history.epoch, history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()

```

```python
learning_curve(history)
```
