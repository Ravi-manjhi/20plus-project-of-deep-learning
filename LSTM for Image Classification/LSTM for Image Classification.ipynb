# Step 01 - Import statments

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras

import matplotlib.pyplot as plt
```

# Step 02- Data preprocessing

```python
from keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# simplifying data

x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
plt.imshow(x_train[0])
```

# Building the model

```python
model = keras.models.Sequential()
```

```python
# adding layers
model.add(keras.layers.LSTM(units=128, activation='relu', input_shape=(28, 28), return_sequences=True))

# droout
model.add(keras.layers.Dropout(0.2))


# adding layers
model.add(keras.layers.LSTM(units=128, activation='relu'))

# droout
model.add(keras.layers.Dropout(0.4))


# fullyConnected layers
model.add(keras.layers.Dense(units=32, activation='relu'))
model.add(keras.layers.Dropout(0.5))

# output layers
model.add(keras.layers.Dense(units=10, activation='softmax'))
```

```python
model.summary()
```

```python
# compile
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
```

```python
# training the mode
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
# evaluation of model
loss, accuracy = model.evaluate(x_train, y_train)
```

```python
print('the Training score:', accuracy)
```

```python
y_pred = model.predict(x_test)
```

```python
pred = np.argmax(y_pred, axis=1)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print('the testing accuracy: ', accuracy_score(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```
