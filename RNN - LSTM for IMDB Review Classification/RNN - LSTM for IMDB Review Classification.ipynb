# Import Statements

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras 
import matplotlib.pyplot as plt
from keras.preprocessing.sequence import pad_sequences
from sklearn.metrics import accuracy_score, confusion_matrix

from keras.datasets import imdb
```

# Data Preprocessing

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=20000)
```

```python
x_train.shape, x_test.shape
```

```python
x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100)
```

```python
x_train.shape, x_test.shape
```

# Building and Evaluation the model

```python
model = keras.models.Sequential()
```

```python
# Adding Embedding layers
model.add(keras.layers.Embedding(input_dim=20000, output_dim=128, input_shape=(100,)))
```

```python
# LSTM layers
model.add(keras.layers.LSTM(units=128, activation='tanh'))

# dropout Layers
model.add(keras.layers.Dropout(0.5))
```

```python
# output layers
model.add(keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# compile the model
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# training the model

history = model.fit(x_train,
                    y_train,
                    batch_size=20,
                    epochs=5,
                    validation_data=(x_test, y_test))
```

```python
# Evaluation 
loss, accuracy = model.evaluate(x_train, y_train)
```

```python
print("The Training datasets score:", accuracy)
```

```python
# Prediction
y_pred = model.predict(x_test)
```

```python
pred = np.where(y_pred > 0.5, 1,0)
```

```python
print(accuracy_score(y_test, pred))
```

```python
confusion_matrix(y_test, pred)
```

```python
def learning_curve(history):
  plt.plot(history.epoch, history.history['accuracy'])
  plt.plot(history.epoch, history.history['val_accuracy'])
  plt.title('Model Accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['train', 'val'], loc='upper left')
  plt.show()

  # loss

  plt.plot(history.epoch, history.history['loss'])
  plt.plot(history.epoch, history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('loss')
  plt.xlabel('Epoch')
  plt.legend(['train', 'val'], loc='upper left')
  plt.show()
```

```python
learning_curve(history)
```
