### converting Greyscale image from RGB Image

1. Using OpenCV
2. using PIL

```python
# Using PIL Module
img = Image.open('./log.jpg').convert('L')
```

```python
img = np.array(img)
img.shape
```

```python
plt.imshow(img, cmap='gray')
```

```python
# using opencv
```

```python
img = cv2.imread('./log.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
gimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gimg, cmap='gray')
```

```python
gimg.shape
```

```python
cv2.imshow('Image', gimg)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
