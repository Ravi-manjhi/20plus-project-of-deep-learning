# Import statements

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
```

# Data Preprocessing

```python
df = pd.read_csv('./GOOG.csv')
```

```python
df.head()
```

```python
closed = df[['Close']].values
closed.shape
```

```python
x = []
y = []

for i in range(closed.shape[0]):
  x.append(closed[i-60:i, 0])
  y.append(closed[i, 0])
```

```python
x = x[60:]
y = y[60:]
```

```python
x = np.array(x)
y = np.array(y).reshape(-1, 1)
```

```python
# StandardScaler
sc = MinMaxScaler()
x = sc.fit_transform(x)
y = sc.fit_transform(y)
```

```python
# splitting the data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=(0.2))
```

```python
x_train.shape, x_test.shape
```

```python
# reshaping the datasets
x_train, x_test = x_train.reshape(959, 60, 1), x_test.reshape(240, 60, 1)
```

# Building the model

```python
model = keras.models.Sequential()
```

```python
# adding Lstm layers
model.add(keras.layers.LSTM(units=32, activation='relu',input_shape=(60, 1),return_sequences=True))

# dropout layers
model.add(keras.layers.Dropout(0.2))

# adding Lstm layers
model.add(keras.layers.LSTM(units=64, activation='relu',input_shape=(60, 1),return_sequences=True))

# dropout layers
model.add(keras.layers.Dropout(0.2))


# adding Lstm layers
model.add(keras.layers.LSTM(units=128, activation='relu',input_shape=(60, 1)))

# dropout layers
model.add(keras.layers.Dropout(0.2))
```

```python
# fullyconnected layers
model.add(keras.layers.Dense(units=128, activation='relu'))

# output layers
model.add(keras.layers.Dense(units=1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# training the model

model.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test))
```

```python
# pred 

pred = model.predict(x_test)
```
